## Filter the WHO List for high confidence matches
- 93% accuracy and above
- Output should be a list of drug bank IDs

```python
import pandas as pd
```

```python
who_matches = pd.read_csv('who_matches.csv')
```

```python
who_matches.head()
```

```python
who_filtered = who_matches[who_matches['score'] >= 93]
```

```python
who_filtered.head(50)
```

```python
# save to csv
who_filtered.to_csv('who_filtered.csv', index=False)
```

```python
drugbank_ids = who_filtered['drugbank_id'].tolist()
```

```python
with open('who_filtered_db_ids.txt', 'w') as f:
    for db_id in drugbank_ids:
        f.write(db_id + '\n')
```
